```python
train_data['A16'] = train_data['A16'].replace({'-': 0, '+': 1})
train_data
```

```python
def impute_nan_with_class_median(data, columns_to_impute, class_column):
    median_per_class = data.groupby(class_column)[columns_to_impute].median()

    for column_name in columns_to_impute:
        for class_value, median_value in median_per_class[column_name].items():
            data.loc[(data[class_column] == class_value) & (data[column_name].isna()), column_name] = median_value

    return data, median_per_class
```

```python
num_columns_to_impute = ['A14', 'A2']
train_data_imputed, numerical_median = impute_nan_with_class_median(train_data, num_columns_to_impute, 'A16')
```

```python
def impute_nan_with_class_mode(data, columns_to_impute, class_column):
    mode_per_class = data.groupby(class_column)[columns_to_impute].apply(lambda x: x.mode().iloc[0])

    for column_name in columns_to_impute:
        for class_value, mode_value in mode_per_class[column_name].items():
            data.loc[(data[class_column] == class_value) & (data[column_name].isna()), column_name] = mode_value

    return data, mode_per_class
```

```python
cat_columns_to_impute = ['A7', 'A6', 'A5', 'A4', 'A1']
train_data_imputed, categorical_mode = impute_nan_with_class_mode(train_data_imputed, cat_columns_to_impute, 'A16')
```

```python
def standarize_and_ohe(data):
  categorical_columns = data.select_dtypes(include=['object']).columns.tolist()
  numerical_columns = data.select_dtypes(include=['number']).columns.tolist()
  numerical_columns.remove('A16')

  numerical_data = data[numerical_columns]
  categorical_data = data[categorical_columns]
  label = data['A16']

  categorical_data = pd.get_dummies(categorical_data)
  numerical_data, standarizer = standarizerData(numerical_data)

  clean_data = pd.concat([numerical_data, categorical_data, label], axis=1)

  return clean_data, standarizer
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
def standarizerData(data):
    data_columns = data.columns
    data_index = data.index

    standardizer = StandardScaler()
    standardizer.fit(data)

    standarized_data_raw = standardizer.transform(data)
    standarized_data = pd.DataFrame(standarized_data_raw)
    standarized_data.columns = data_columns
    standarized_data.index = data_index

    return standarized_data, standardizer
```

```python
train_data_clean, standarizer = standarize_and_ohe(train_data_imputed)
train_data_clean
```

```python
def fill_nan_test_data(data, value_df, class_column, columns_to_impute):
    for column_name in columns_to_impute:
        for class_value in data[class_column].unique():
            impute_value = value_df.loc[class_value, column_name]
            data.loc[(data[class_column] == class_value) & (data[column_name].isna()), column_name] = impute_value
    return data
```

```python
test_data_imputed = fill_nan_test_data(test_data, numerical_median, 'A16', num_columns_to_impute)
test_data_imputed = fill_nan_test_data(test_data_imputed, categorical_mode, 'A16', cat_columns_to_impute)
test_data_imputed
```
